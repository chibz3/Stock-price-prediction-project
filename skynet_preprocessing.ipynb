```python
subreddit = 'teslamotors'
limit = 100
timeframe = 'day' #hour, day, week, month, year, all
listing = 'new' # controversial, best, hot, new, random, rising, top
  
def get_reddit(subreddit,listing,limit,timeframe):
    try:
        base_url = f'https://www.reddit.com/r/{subreddit}/{listing}.json?limit={limit}&t={timeframe}'
        request = requests.get(base_url, headers = {'User-agent': 'yourbot'})
    except:
        print('An Error Occured')
    return request.json()
 
def get_post_titles(r):
    '''
    Get a List of post titles
    '''
    posts = []
    for post in r['data']['children']:
        x = post['data']['title']
        posts.append(x)
    return posts
 
def get_results(r):
    '''
    Create a DataFrame Showing Title, URL, Score and Number of Comments.
    '''
    myDict = {}
    for post in r['data']['children']:
        myDict[post['data']['title']] = {'url':post['data']['url'],'score':post['data']['score'],'comments':post['data']['num_comments']}
    df = pd.DataFrame.from_dict(myDict, orient='index')
    return df
 
if __name__ == '__main__':
    r = get_reddit(subreddit,listing,limit,timeframe)
    df = get_results(r)
```

### Pre-Process

```python
reddit_data = df.copy()
```

```python
reddit_data.reset_index(level = 0, inplace = True)
```

```python
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
```

```python
for i, date in enumerate(dates):
  numerics = int(date.split(' ')[0])
  if (date[-9:] == 'minute ago') | (date[-11:] == 'minutes ago'):
    dates[i] = today
  elif (date[-9:] == 'hours ago') | (date[-8:] == 'hour ago'):
    if numerics < int(current_time[:2]):
      dates[i] = today
    else:
      dates[i] = today - timedelta(days = 1)
  else:
    dates[i] = today - timedelta(days = numerics)
```

```python
dates = [date for date in dates if date == today]
```

```python
for i, date in enumerate(dates):
  output = str(date.year)
  if len(str(date.month)) < 2:
    output += '-0{}'.format(date.month)
  else:
    output += '-{}'.format(date.month)
  if len(str(date.day)) < 2:
    output += '-0{}'.format(date.day)
  else:
    output += '-{}'.format(date.day)
  dates[i] = output
```

```python
reddit_data = reddit_data.head(len(dates))
```

```python
reddit_data['date'] = dates
```

```python
reddit_data.drop(columns=['url', 'score', 'comments'],inplace=True)
reddit_data.rename({'index':'text'},axis=1,inplace=True)
reddit_data['text'] = reddit_data.text.apply(preprocess)
```
