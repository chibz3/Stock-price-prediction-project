```python
MAX_LENGTH_NEWS = 150
MAX_LENGTH_REDDIT = 70
MAX_LENGTH_TWITTER = 60
```

```python
# Output sentiment
tokens_Fin=tokenizer_Fin(news_df.text.to_list(),padding=True,truncation=True,max_length=MAX_LENGTH_NEWS,return_tensors="pt")
outputs_Fin=model_Fin(**tokens_Fin)
sentiment = F.softmax(outputs_Fin.logits, dim=-1).detach().numpy()
```

```python
# Difference positive - negative sentiment
sentiment = sentiment[:,0]- sentiment[:,1]
```

```python
news_ts = pd.DataFrame({'date' : news_df.date,'sentiment' : sentiment})
```

```python
news_ts['date'] = news_ts.date.astype('datetime64')
```

```python
# fill lacking days with movin average
r = pd.date_range(start=news_ts.date.min(), end=news_ts.date.max())
news_ts=news_ts.groupby('date').mean().reindex(r)
fill=news_ts.sentiment.rolling(3, center=True,min_periods=1).mean()
news_ts=news_ts.sentiment.fillna(fill).reset_index().rename({'index':'date'},axis=1)
```

## Formatting Data

```python
current_day = pd.to_datetime(today,format='%Y-%m-%d')
```

```python
data = yf.Ticker('TSLA')
data = data.history(start= current_day - pd.to_timedelta(25, unit='d'),end = current_day, interval='1d', rounding=True)
data = data[['Close']].reset_index()
```

```python
r = pd.date_range(start=data.Date.min(), end=pd.to_datetime(data.Date.max(), format='%Y-%m-%d'))
data=data.groupby('Date').mean().reindex(r)
data.fillna(method='ffill',inplace=True)
data = data.reset_index().rename({'index':'Date'},axis=1)
data = data.loc[data.Date >= current_day - pd.to_timedelta(21, unit='d')].reset_index(drop=True).rename({'Date':'date'},axis=1)
```

```python
def scale_tesla(data):
  return (data - 550)/(900-550)
def scale_tesla_inv(data):
  return data*(900-550) + 550
```

```python
data['Close'] = data.Close.apply(scale_tesla)
```

## Merging Sentiments

```python
obj = s3_client.get_object(Bucket='skynet-brain', Key= 'dp_prepro_models/GRU_model_weights.h5')
elon = pd.read_csv(obj['Body'], index_col=0)
elon['date'] = pd.to_datetime(elon.date,format='%Y-%m-%d')
```

```python
obj = s3_client.get_object(Bucket='skynet-brain', Key= 'dp_prepro_models/GRU_model_weights.h5')
reddit = pd.read_csv(obj['Body'], index_col=0)
reddit.fillna(0,inplace=True)
reddit['date'] = pd.to_datetime(reddit.date,format='%Y-%m-%d')
```

```python
obj = s3_client.get_object(Bucket='skynet-brain', Key= 'dp_prepro_models/GRU_model_weights.h5')
news = pd.read_csv(obj['Body'], index_col=0)
news['date'] = pd.to_datetime(news.date,format='%d/%m/%Y')
```

```python
data = data.merge(elon,on='date').merge(reddit,on='date').merge(news,on='date')
```

## Model

```python
import tensorflow as tf
```

```python
dataset = tf.data.Dataset.from_tensor_slices(data.iloc[:,1:])

dataset = dataset.window(21, shift=1, drop_remainder=True)

dataset = dataset.flat_map(lambda w: w.batch(21))

dataset = dataset.map(lambda w: (w))

dataset = dataset.take(1).batch(8).prefetch(1)
```

```python
import tensorflow.keras as keras
from tensorflow.keras import Sequential
from keras.layers import Input, Flatten, Dense, Dropout, LSTM, GRU
from tensorflow.keras import callbacks
```

```python
model = Sequential([
    Input(shape=(21,4)),
    #Model layers
    GRU(80, return_sequences=True),
    Flatten(),
    Dense(600, activation="relu"),
    Dense(120, activation="relu"),
    Dense(16, activation="relu"),
    Dense(7)
])

model.load_weights(gru_weights)
```

```python
predictions = model.predict(dataset)
predictions = scale_tesla_inv(predictions)
```

```python
true = yf.Ticker('TSLA')
true = true.history(start= current_day ,end = current_day + pd.to_timedelta(8, unit='d'), interval='1d', rounding=True)
true = true[['Close']].reset_index()
r = pd.date_range(start=current_day, end=current_day+pd.to_timedelta(6, unit='d'))
true=true.groupby('Date').mean().reindex(r)
```

```python
true['pred'] = predictions[0,:]
predictions_NN = true.dropna().pred.values
```
